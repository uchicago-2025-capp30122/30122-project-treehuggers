```python
from keplergl import KeplerGl
from pathlib import Path
import json
import geopandas as gpd
import os
```

```python
#data_paths
data_parent = Path(os.getcwd()).parent
print(f"data parent{data_parent}")
path_parks = data_parent / "data/cleaned_park_polygons.geojson"
path_housing = data_parent / "data/housing_data_index.geojson"
path_tracts = data_parent / "data/grid_and_tracts/processed/merged/merged_tract_data.geojson"
output_file = data_parent / "scripts/viz/chicago_parks_kepler.html"
```

```python
def load_geojson_data(file_path):
    """Load GeoJSON files."""
    try:
        gdf = gpd.read_file(file_path)
        for col in gdf.columns:
            # Check if the column contains complex types
            if gdf[col].dtype == 'object':
                print(f"Had to convert column{col}")
                # convert to string
                gdf[col] = gdf[col].fillna('').astype(str)
        return gdf
    except Exception as e:
        print(f"Error loading GeoJSON file {file_path}: {e}")
        return None
```

```python
with open(path_parks) as f:
    parks_data = json.load(f)
    
with open(path_housing) as f:
    housing_data = json.load(f)

with open(path_tracts) as f:
    tracts_data = json.load(f)
```

```python
with open("kepler_config.json", "r") as f:
    config = json.load(f)
```

```python
kepler_map = KeplerGl(height=800, width=1200, 
                      data={
                          'Housing': housing_data, 
                          'Parks': parks_data, 
                          'Tracts': tracts_data
                      }, config=config)
kepler_map
```

```python
config_2 = kepler_map.config

with open("kepler_config2.json", "w") as f:
    json.dump(config_2, f)
```
